```python
import numpy as np
import pandas as pd
import seaborn as sb
import seaborn as sns
import matplotlib.pyplot as plt # we only need pyplot
sb.set()
```

```python
spotifyData = pd.read_csv('spotify-2023.csv', encoding='ISO-8859-1')
spotifyData.head()
```

```python
spotifyData.info()
```

```python
#dropping track_name, artist_count, released year, month and day. I feel like it doesn't really affect the rankings of the songs to a very large extent though.. not too sure we can discuss this later
```

```python
clean_data = spotifyData.drop(columns = ['track_name', 'artist_count', 'released_year', 'released_month', 'released_day'])
```

```python
clean_data.info()
```

```python
#checking for any NaN values
```

```python
clean_data.isnull().sum()
```

```python
cleaner_data = clean_data.dropna(subset = ['in_shazam_charts', 'key'])
cleaner_data.info()
```

```python
cleaner_data.isnull().sum()
```

```python
#no more NaN values
```

```python
#remove commas for data in in_deezer_playlists and in_shazam_charts
```

```python
cleaner_data.loc[:,'in_shazam_charts'] = cleaner_data['in_shazam_charts'].astype(str).str.replace(',', '')

cleaner_data.loc[:,'in_shazam_charts'] = pd.to_numeric(cleaner_data['in_shazam_charts'], errors='coerce')

cleaner_data.loc[:,'in_deezer_playlists'] = cleaner_data['in_deezer_playlists'].astype(str).str.replace(',', '')

cleaner_data.loc[:,'in_deezer_playlists'] = pd.to_numeric(cleaner_data['in_deezer_playlists'], errors='coerce')
```

```python
cleaner_data.info()
```

```python
#converting data eg streams back to int cause its objects now
```

```python
cleaner_data.loc[:,'streams'] = pd.to_numeric(cleaner_data['streams'], errors='coerce')
cleaner_data.loc[:,'in_deezer_playlists'] = pd.to_numeric(cleaner_data['in_deezer_playlists'], errors='coerce')
cleaner_data.loc[:,'in_shazam_charts'] = pd.to_numeric(cleaner_data['in_shazam_charts'], errors='coerce')
```

```python
cleaner_data.info()
```

```python
df_encoded = pd.get_dummies(cleaner_data, columns=['artist(s)_name', 'key', 'mode'])
df_encoded
```

```python
#replacing true and false values with binary data instead
```

```python
df_encoded.mode_Major = df_encoded.mode_Major.replace({True: 1, False: 0})
df_encoded
```

```python
df_encoded.mode_Minor = df_encoded.mode_Minor.replace({True: 1, False: 0})
df_encoded
```

```python
df_encoded.key_D = df_encoded.key_D.replace({True: 1, False: 0})
df_encoded.key_E = df_encoded.key_E.replace({True: 1, False: 0})
df_encoded.key_F = df_encoded.key_F.replace({True: 1, False: 0})
df_encoded.key_G = df_encoded.key_G.replace({True: 1, False: 0})
df_encoded['key_C#'] = df_encoded['key_C#'].replace({True: 1, False: 0})
df_encoded['key_D#'] = df_encoded['key_D#'].replace({True: 1, False: 0})
df_encoded['key_F#'] = df_encoded['key_F#'].replace({True: 1, False: 0})
df_encoded['key_G#'] = df_encoded['key_G#'].replace({True: 1, False: 0})
df_encoded
```

```python
data = df_encoded
data.info()
```

```python
combined_df = pd.concat([cleaner_data, data], axis=1)
```
